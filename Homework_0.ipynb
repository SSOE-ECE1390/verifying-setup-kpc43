* Load an (Grayscale) image  (‘Data/Apollo_11_Launch.jpg’)
* View image
* Crop image
* Edit brightness of image
* Rotate image
* Save modified image


```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("Data/Apollo_11_Launch.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.show()

crop = img[100:400, 200:500]

plt.imshow(crop, cmap="gray")
plt.title("Cropped Image")
plt.axis("off")
plt.show()

bright = cv2.convertScaleAbs(img, alpha=1, beta=50)

plt.imshow(bright, cmap="gray")
plt.title("Brightness Adjusted")
plt.axis("off")
plt.show()

rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(rotated, cmap="gray")
plt.title("Rotated Image")
plt.axis("off")
plt.show()


cv2.imwrite("Data/Apollo_11_Modified.jpg", rotated)
print("Modified image saved as Data/Apollo_11_Modified.jpg")
```
